```python
import tensorflow as tf

from sklearn.datasets import load_digits
digits = load_digits()
XX = digits.data    # 数字图片特征提取后得到的向量
y = digits.target  #  y是数字，如 0，1，2，..., 9
from OneHot import OneHot
yy=OneHot(y)
aver=XX.mean(axis=0)
std=XX.std(axis=0)
std[std==0]=1
XX=(XX-aver)/std
```

```python
batch_size = 5 # 每次修正模型使用的样本个数
tf.compat.v1.disable_eager_execution()
x = tf.compat.v1. placeholder(tf.float32, [None, 64])  # 64个输入节点
y = tf.compat.v1.placeholder(tf.float32, [None, 10])  # 10个输出节点
W = tf.compat.v1.Variable(tf.zeros([64, 10])) # 定义模型权重
b = tf.compat.v1.Variable(tf.zeros([10]))  # 定义
# 用softmax构建 逻辑回归模型 
pred = tf.nn.softmax(tf.matmul(x, W) + b)  
# 损失函数(用交叉熵)
cost = tf.reduce_mean(-tf.reduce_sum(y*tf.math.log(pred), 1)) 
#  低度下降算法—梯度下降
optimizer = tf.compat.v1.train.GradientDescentOptimizer(0.01).minimize(cost)
# 初始化所有变量
init = tf.compat.v1.global_variables_initializer()
```

```python
with tf.compat.v1.Session() as sess:
    sess.run(init)
    for epoch in range(25): # 开始训练  25次
        avg_cost = 0.
        total_batch = len(XX)//batch_size
        for i in range(total_batch):
            batch_xs=XX[i*batch_size:(i+1)*batch_size]
            batch_ys=yy[i*batch_size:(i+1)*batch_size]
            sess.run(optimizer, {x: batch_xs,y: batch_ys})
            #计算损失平均值
            avg_cost += sess.run(cost,{x: batch_xs,y: batch_ys}) / total_batch
        if (epoch+1) % 5 == 0:
            print("Epoch:", '%04d' % (epoch+1), "cost=", "{:.9f}".format(avg_cost))
    print("建模完成")
```

```python
from sklearn.datasets import make_circles
import numpy as np
X,y = make_circles(100, factor=.1, noise=.1)  # 产生模拟数据
# 模拟的数据，如图所示
from OneHot import OneHot
y = OneHot(y)  # y原来是整数0或1，变成0=1 0，1=0 1
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=.1) 


# 1  引用
from keras.models import Sequential   # 网络模型
from keras.layers import Dense   #  网络的一个层
# 2 生成模型
model = Sequential()
# 3 增加网络层：unit指定节点数， activation激活函数
Layer1= Dense(units=30, activation='relu',input_dim=2) 
Layer2= Dense(units=2, activation='softmax') 
model.add(Layer1) 
model.add(Layer2) 
#输入层需要input_dim参数，较特殊, units则是指input的下一层
# 4 编译模型：指定优化、损失函数
model.compile(loss='categorical_crossentropy', optimizer='sgd', metrics=['accuracy'])
# 5 训练和预报：指定训练集，每次循环修正模型的样本数
history = model.fit(Xtrain, ytrain,validation_data= (Xtest, ytest), epochs=200, verbose = 1) 
# classes = model.predict(Xtest, batch_size=5)





import matplotlib.pyplot as plt
plt.plot(history.history['accuracy'],label = 'train')
plt.plot(history.history['val_accuracy'],label = 'test')
plt.legend()

classes = model.predict(Xtest, batch_size=5)
Ypred = classes.argmax(axis=1)
Ytrue = np.argmax(ytest,axis=1)
print('pred',Ypred)
print('true',Ytrue)

model.save('circleClass.bin')
```
